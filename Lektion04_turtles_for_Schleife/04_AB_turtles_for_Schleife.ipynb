## Wichtige Punkte zur `for`-Schleife

1. **Die `for`-Schleife ist eine spezielle Art von Schleife.**  
   Eine `for`-Schleife ist eine Kontrollstruktur in der Informatik, die es erm√∂glicht, einen Codeblock wiederholt auszuf√ºhren.  
   Sie wird h√§ufig verwendet, um √ºber eine Sequenz von Elementen ‚Äì zum Beispiel eine Zahlenreihe ‚Äì zu **iterieren** (d. h. wiederholen).  
   Dabei wird ein **Z√§hler** verwendet, der bei jeder Iteration der Schleife einen neuen Wert annimmt.

2. **Die Funktion `range()` legt fest, wie oft die Anweisung wiederholt wird.**  
   Beispielsweise bedeutet `range(4)` eine viermalige Wiederholung und `range(10)` eine zehnmalige Wiederholung.  
   Tats√§chlich generiert `range(4)` die vier Zahlen `0`, `1`, `2` und `3`, und der Z√§hler nimmt nacheinander jeden dieser Werte an.  
   Da es vier Zahlen gibt, die den Wert annehmen k√∂nnen, finden **vier Wiederholungen** statt.


## Vertiefung: Die `for`-Schleife mit einer Zahlenfolge benutzen

Die `for`-Schleife funktioniert mit Hilfe der `range()`-Funktion auch mit Zahlenbereichen.  
Nehmen Sie die Zahl `2` und multiplizieren Sie sie f√ºnfmal mit der Zahl `3`.

```python
zahl = 2
for i in range(0, 5):
    zahl = zahl * 3
print("Das Ergebnis lautet", zahl)
```

In diesem Beispiel wird der Wert der Variablen `zahl` in jedem Schleifendurchlauf mit `3` multipliziert.  
Da `range(0, 5)` f√ºnf Werte liefert (`0, 1, 2, 3, 4`), wird die Anweisung **f√ºnfmal** ausgef√ºhrt.  
Nach Abschluss der Schleife enth√§lt `zahl` somit das Ergebnis der f√ºnffachen Multiplikation mit `3`.

---

üîπ **WICHTIG:** Die `range()`-Funktion inkludiert nur den unteren Wert, **nicht den oberen**.  
Beispiel: `range(0, 5)` erzeugt **0 bis 4**, aber **nicht 5**.  

üîπ **Die Schleife l√§uft 5 Durchl√§ufe:** den **0.**, **1.**, **2.**, **3.** und **4.**
