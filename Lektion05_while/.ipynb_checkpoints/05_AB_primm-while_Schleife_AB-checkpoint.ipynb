## INVESTIGATE – Begriffe und Beispiele zu Schleifen

In diesem Abschnitt klären wir die wichtigsten Begriffe rund um Schleifen in Python.

### Schlüsselbegriffe – Lektion 5 Schleifen

| Begriff     | Bedeutung                                                                 | Wie man es in Python macht                                         |
|------------|----------------------------------------------------------------------------|---------------------------------------------------------------------|
| Iteration  | Eine **Wiederholung** in einer Schleife.                                   | Tritt in `for`- und `while`-Schleifen auf.                         |
| for-Schleife | Schleife mit **fester Anzahl** an Wiederholungen.                        | `for zahl in range(1, 11):`<br>`    print(zahl)`                   |
| while-Schleife | **Bedingte Schleife**: läuft so lange weiter, wie eine Bedingung wahr ist. | `while zahl < 11:`<br>`    print(zahl)`<br>`    zahl = zahl + 1`    |

#### Vergleich: for-Schleife und while-Schleife (Beispiel Hochzählen 1 bis 10)

```python
# for-Schleife
for zahl in range(1, 11):
    print(zahl)

# while-Schleife
zahl = 1
while zahl < 11:
    print(zahl)
    zahl = zahl + 1
```

**Merken Sie sich:**

- Immer wenn ein **fest definierter Bereich** durchlaufen werden soll, ist die `for`-Schleife meist einfacher zu verwenden.
- Die `while`-Schleife ist sinnvoll, wenn die Anzahl der Wiederholungen **nicht von vornherein feststeht**, sondern von einer **Bedingung** abhängt (z.B. Benutzereingaben).

### Beispiel: flexible Anzahl von Schleifendurchläufen

```python
zahl = 0
fortfahren = "j"

while fortfahren == "j":
    zahl = zahl + 1
    print(zahl)
    fortfahren = input("Geben Sie 'j' ein, wenn Sie fortfahren möchten: ")

print(f"Sie haben bis {zahl} gezählt.")
```

Hier bestimmt der **Nutzer**, wie oft die Schleife durchlaufen wird.

### Vorsicht: Endlosschleife

Sie haben voller Freude ein neues Programm geschrieben, drücken auf „Ausführen“ und warten ... und warten ... und warten!  
Wenn das passiert, sind Sie möglicherweise in einer **Endlosschleife** gefangen.

Eine Endlosschleife ist eine Schleife, die ständig wiederholt wird, ohne jemals zu Ende zu kommen.

```python
zahl = 1
while zahl < 11:
    print(zahl)
    # zahl = zahl + 1   # Hochzählen vergessen – die Bedingung bleibt immer wahr!
```

Endlosschleifen sind **logische Fehler** und müssen behoben werden, damit das Programm tut, was es soll.

Bei jedem Durchlauf soll ausgegeben werden:

````
# Aufgabe 9 – Lösungsidee mit for-Schleife
anzahl = int(input("Wie oft soll die Schleife durchlaufen werden? "))

for durchlauf in range(1, anzahl + 1):
    print(f"Dies ist der {durchlauf}. Schleifendurchlauf.")
````

### Aufgabe 12: Kalorienbilanz über 3 Tage

Ihr Kalorienbedarf liegt bei ca. **2200 kcal pro Tag**.  
Sie beschließen, eine Diät zu machen und für **3 Tage** weniger zu essen.

Das Programm soll Folgendes tun:

1. Das Programm fragt Sie, **wie viele Mahlzeiten** Sie an einem Tag zu sich nehmen.
2. Dann fragt es für **jeden Tag** und für **jede Mahlzeit** nach den Kalorien, die Sie zu sich genommen haben.
3. Am Ende zieht das Programm eine **Gesamtbilanz**.
4. **7000 kcal** entsprechen ungefähr **1 kg Körperfett**.  
   Das Programm sagt Ihnen, wie viele kg Fett Sie mit dieser Aktion **verloren** oder ggf. **zugenommen** haben.

*Hinweis:* Arbeiten Sie mit **verschachtelten Schleifen** (Schleife über Tage, Schleife über Mahlzeiten).

````
# Aufgabe 12 – Starter-Code
tage = 3
grundbedarf_pro_tag = 2200

gesamt_kalorien = 0

mahlzeiten_pro_tag = int(input("Wie viele Mahlzeiten essen Sie pro Tag? "))

for tag in range(1, tage + 1):
    print(f"\nTag {tag}:")
    for mahlzeit in range(1, mahlzeiten_pro_tag + 1):
        kcal = int(input(f"  Kalorien für Mahlzeit {mahlzeit}: "))
        gesamt_kalorien = gesamt_kalorien + kcal

print("\nAuswertung:")
print("Gesamtkalorien in 3 Tagen:", gesamt_kalorien)

# Ergänzen Sie nun eine Berechnung, wie viele kg Körperfett
# Sie im Vergleich zu Ihrem Bedarf (2200 kcal pro Tag) verloren oder zugenommen haben.
````
